```python
# It looks like you’re working with a Pandas DataFrame and trying to find the indices of rows 
# where the “RainTomorrow” column has missing values.
a=[(i,index) for i,index in enumerate(df["RainTomorrow"]) if pd.isna(index)]
result = pd.DataFrame(a).set_index(0)
```

```python
#Creating list of all non null values in the target variable for spliting in train and test data.
a=[]
for i in range(df1.shape[0]):
  if i not in index_nan:
    a.append(i)
```

```python
df1_training_data=df1.iloc[a]
df1_training_data
```

```python
X_training_data=df1_training_data.drop("RainTomorrow",axis=1)
y_training_data=df1_training_data["RainTomorrow"]
```

```python
X_training_data=X_training_data.fillna(method="ffill",axis=0)
X_training_data=X_training_data.fillna(method="bfill",axis=0)
```

```python
from sklearn import preprocessing
L_encoding=preprocessing.LabelEncoder()
X_training_data["Location"]=L_encoding.fit_transform(X_training_data["Location"])
```

```python
X_training_data["WindGustDir"]=L_encoding.fit_transform(X_training_data["WindGustDir"])
```

```python
X_training_data["WindDir9am"]=L_encoding.fit_transform(X_training_data["WindDir9am"])
```

```python
X_training_data["WindDir3pm"]=L_encoding.fit_transform(X_training_data["WindDir3pm"])
```

```python
X_training_data["RainToday"]=L_encoding.fit_transform(X_training_data["RainToday"])
```

```python
X_training_data.head()
```

```python
from sklearn.preprocessing import StandardScaler
scalar=StandardScaler()
x_training1=scalar.fit_transform(X_training_data)
```

```python
x_training2=pd.DataFrame(x_training1)
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(x_training2,y_training_data,test_size=0.20,random_state=0)
```

```python
from sklearn.linear_model import LogisticRegression
logreg=LogisticRegression(solver='liblinear', random_state=0)
logreg.fit(X_train,y_train)
```

```python
y_pred_train_training2=logreg.predict(X_train)
y_pred_train_training2
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_train,y_pred_train_training2)
```

```python
y_pred_test_training2=logreg.predict(X_test)
```

```python
accuracy_score(y_test,y_pred_test_training2)
```
